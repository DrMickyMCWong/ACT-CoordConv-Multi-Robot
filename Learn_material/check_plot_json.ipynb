```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Load the JSON data
file_path = r"C:\Users\Administrator\Documents\Simple_CNN_MLP\datasets\dual_camera_robot_dataset_v2_0618\episode_003\episode_data.json"

with open(file_path, 'r') as f:
    data = json.load(f)

# Extract data
frame_indices = []
leader_b_values = []
follower_b_values = []
episode_times = []

for entry in data:
    frame_indices.append(entry['frame_index'])
    leader_b_values.append(entry['leader_state']['b'])
    follower_b_values.append(entry['follower_state']['b'])
    episode_times.append(entry['episode_time'])

# Create two separate plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Leader state "b" variable
ax1.plot(episode_times, leader_b_values, 'b-', linewidth=2, label='Leader "b"')
ax1.set_xlabel('Episode Time (seconds)')
ax1.set_ylabel('Leader "b" Value')
ax1.set_title('Leader State "b" Variable Over Time')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Plot 2: Follower state "b" variable
ax2.plot(episode_times, follower_b_values, 'r-', linewidth=2, label='Follower "b"')
ax2.set_xlabel('Episode Time (seconds)')
ax2.set_ylabel('Follower "b" Value')
ax2.set_title('Follower State "b" Variable Over Time')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()

# Optional: Create a combined plot for comparison
plt.figure(figsize=(12, 6))
plt.plot(episode_times, leader_b_values, 'b-', linewidth=2, label='Leader "b"', alpha=0.8)
plt.plot(episode_times, follower_b_values, 'r-', linewidth=2, label='Follower "b"', alpha=0.8)
plt.xlabel('Episode Time (seconds)')
plt.ylabel('"b" Value')
plt.title('Leader vs Follower "b" Variable Over Time')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Print some basic statistics
print(f"Episode duration: {max(episode_times):.2f} seconds")
print(f"Total frames: {len(data)}")
print(f"Leader 'b' range: {min(leader_b_values):.6f} to {max(leader_b_values):.6f}")
print(f"Follower 'b' range: {min(follower_b_values):.6f} to {max(follower_b_values):.6f}")
```
